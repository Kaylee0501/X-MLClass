```python
import os
import numpy as np
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID" 
os.environ["CUDA_VISIBLE_DEVICES"] = "3" 
```

```python
with open("../datasets/AAPD/llm_cluster_result/zero_shot_text_test1.json","r") as json_file:
    json_list = list(json_file)
```

```python
predict_output = []
for i in range(len(json_list)):
    predict_output.append(json_list[i].split('"labels": [')[1].split(',')[0][1:-1])
```

```python
file1 = open('../datasets/AAPD/true_labels.txt', 'r')
summary = file1.readlines()  
```

```python
label_dic = {}
for row in summary:
    label_dic[row.split('; ')[0]] = row.split('; ')[1]
```

```python
file2 = open('../datasets/AAPD/label_test.txt', 'r')
org_true_label = file2.readlines()
true_label = []
for row in org_true_label:
    short_label = row.strip().split(' ')
    long_label = []
    for index in short_label:
        long_label.append(label_dic[index])
    true_label.append(long_label)
true_label
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2', device = 0)
```

```python
correct = 0
for index in range(len(predict_output)):
    query_embedding = model.encode(predict_output[index])
    print(predict_output[index], true_label[index])
    passage_embedding = model.encode(true_label[index])
    sim_scores = util.dot_score(query_embedding, passage_embedding)[0].numpy()
    print(sim_scores)
    rank_list = np.argsort(sim_scores)[-1]
    if sim_scores[rank_list] >= 0.60:
        correct +=1
correct
```

```python
ground_truth = []
for row in summary:
    ground_truth.append(row.split('; ')[1])
```

```python
file1 = open('../datasets/AAPD/llm_cluster_result/update_labelspace.txt', 'r')
documents = file1.readlines()  

label_space = []
for row in documents:
    label = row.strip()
    label_space.append(label)
```

```python
correct = 0
for lb in ground_truth:
    query_embedding = model.encode(lb)
    print(lb)
    passage_embedding = model.encode(label_space)
    sim_scores = util.dot_score(query_embedding, passage_embedding)[0].numpy()
    rank_list = np.argsort(sim_scores)[-1]
    print(label_space[rank_list], sim_scores[rank_list])
    if sim_scores[rank_list] >= 0.60:
        correct +=1
correct
```

```python
correct/ len(ground_truth)
```
